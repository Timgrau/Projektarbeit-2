```python
from qutip import Bloch, Qobj, states
import numpy as np
import matplotlib.pyplot as plt
```

```python
bloch_0 = Bloch()
bloch_1 = Bloch()

base_0 = states.basis(2, 0)
base_1 = states.basis(2, 1)

bloch_0.add_states(base_0)
bloch_1.add_states(base_1)

# Blochs-Kugel settings
bloch_0.sphere_alpha = 0.05
bloch_0.sphere_color = "w"
bloch_0.xlabel = ['$\\left|+\\right>$', "$\\left|-\\right>$"]  
bloch_0.view = [-50, 30]

# Farben und Markierungen
bloch_0.vector_color = ["r"]
bloch_0.point_size = [45, 70, 70]

# Blochs-Kugel settings
bloch_1.sphere_alpha = 0.05
bloch_1.sphere_color = "w"
bloch_1.xlabel = ['$\\left|+\\right>$', "$\\left|-\\right>$"]  
bloch_1.view = [-50, 30]

# Farben und Markierungen
bloch_1.vector_color = ["r"]
bloch_1.point_size = [45, 70, 70]
```

```python
bloch_0.show()
bloch_1.show()
```
